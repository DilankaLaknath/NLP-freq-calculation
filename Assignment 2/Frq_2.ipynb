```python
import csv
import re

# Create a dictionary to store the word frequencies
word_freq = {}
```

```python
# Read in the text file
with open('28757_Tamil.txt', 'r', encoding='utf-8') as f:
    for line in f:
        # Split the line into words, removing leading/trailing whitespace and punctuation
        words = [word.strip(' ?:.,;-()/"_') for word in line.split()]
        # Add each word to the dictionary or increment its count
        for word in words:
            word_freq[word] = word_freq.get(word, 0) + 1
```

```python
# Sort the dictionary by value (i.e., by frequency) in descending order
word_freq = {k: v for k, v in sorted(word_freq.items(), key=lambda item: item[1], reverse=True)}
```

```python
# Create a new dictionary excluding numeric and space keys
word_freq = {k: v for k, v in word_freq.items() if not (k.isdigit() or k == '')}
```

```python
# Write the results to a CSV file
with open('Frq_2.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    # Write the header row
    writer.writerow(['Word', 'Freq'])
    # Write each word-frequency pair to a separate row
    for word, freq in word_freq.items():
        writer.writerow([word, freq])
```

```python
# Create a dictionary to store the frequency of frequencies
freq_freq = {}
```

```python
# Count the frequency of each frequency
for freq in word_freq.values():
    freq_freq[freq] = freq_freq.get(freq, 0) + 1
```

```python
# Sort the dictionary by value (i.e., by frequency) in descending order
freq_freq = {k: v for k, v in sorted(freq_freq.items(), key=lambda item: item[1], reverse=True)}
```

```python
# Write the results to a CSV file
with open('FrFr_2.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    # Write the header row
    writer.writerow(['Freq', 'FreqOfFreq'])
    # Write each frequency-frequency pair to a separate row
    for freq, freq_of_freq in freq_freq.items():
        writer.writerow([freq, freq_of_freq])
```

```python
# Rank the words by frequency
ranked_words = [(word, freq, rank) for rank, (word, freq) in enumerate(word_freq.items(), start=1)]
```

```python
# Write the results to a CSV file
with open('Rank_2.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    # Write the header row
    writer.writerow(['Word', 'Freq', 'Rank'])
    # Write each word-frequency-rank triple to a separate row
    for word, freq, rank in ranked_words:
        writer.writerow([word, freq, rank])
```

```python
import matplotlib.pyplot as plt
import csv

# Create lists to store the word frequencies and ranks
word_freq = []
ranks = []

# Read in the CSV file containing the word frequencies and ranks
with open('Rank_2.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        word_freq.append(int(row['Freq']))
        ranks.append(int(row['Rank']))

# Plot the rank vs frequency graph
plt.scatter(ranks, word_freq)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.title('Zipf\'s Law')
plt.grid(True)

# Save the graph to a PNG file
plt.savefig('Graph_2.png')
```

```python
import csv
import math

# Read in the CSV file containing the word frequencies
with open('Frq_2.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    word_freq = {row['Word']: int(row['Freq']) for row in reader}

# Calculate the mean and standard deviation of the word frequencies
freq_values = list(word_freq.values())
mean_freq = sum(freq_values) / len(freq_values)
stddev_freq = math.sqrt(sum([(f - mean_freq)**2 for f in freq_values]) / len(freq_values))

# Calculate the z-score of each word frequency and store it in a new dictionary
word_zscores = {}
for word, freq in word_freq.items():
    z = (freq - mean_freq) / stddev_freq
    word_zscores[word] = z

# Write the results to a CSV file
with open('Z_2.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    # Write the header row
    writer.writerow(['Word', 'Freq', 'Z'])
    # Write each word-frequency-zscore triple to a separate row
    for word, freq in word_freq.items():
        zscore = word_zscores[word]
        writer.writerow([word, freq, zscore])
```

```python
import csv

# Set the z-score threshold for stop words
z_threshold = 1.96

# Read in the CSV file containing the z-scores
with open('Z_2.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    word_zscores = {row['Word']: float(row['Z']) for row in reader}

# Identify the stop words based on the z-score threshold
stop_words = [word for word, zscore in word_zscores.items() if zscore >= z_threshold]

# Write the stop words to a text file
with open('Stop_2.txt', 'w', encoding='utf-8') as f:
    f.write('\n'.join(stop_words))
```
